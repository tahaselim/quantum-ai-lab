# Quantum Entanglement
$ \newcommand{\ket}[1]{|#1\rangle} $

Entanglement quantum circuit:



| **Input State $\ket{ab}$**  | **Output State(Bell State)**            |
|-----------------------------|-----------------------------------------|
|$\ket{00}$ | $ \ket{\phi^+} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$ |
|$\ket{01}$ | $ \ket{\psi^+} = \frac{1}{\sqrt{2}}(\ket{01} + \ket{10})$ |
|$\ket{10}$ | $ \ket{\phi^-} = \frac{1}{\sqrt{2}}(\ket{00} - \ket{11})$ |
|$\ket{11}$ | $ \ket{\psi^-} = \frac{1}{\sqrt{2}}(\ket{01} - \ket{10})$ |

The output qubits are correlated to each other >> **Entangled state**.

For the state $ \ket{\phi^+} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$:\
We can observe either state $ \ket{00} $ or state $ \ket{11} $, if meaused the 2 qubits.

If Alice observes her qubit secretly:
<ul>
    <li> When Alice sees the result $ \ket{0} $, then Bob's qubit also collapses to state $ \ket{0} $. Bob cannot observe state $ \ket{1} $. </li>
    <li> When Alice sees the result $ \ket{1} $, then Bobs's qubit also collapses to state $ \ket{1} $. Bob cannot observe state $ \ket{0} $. </li>
</ul>

Experimental results have confirmed that this happens even if there is a physical distance between Alice's and Bob's qubits. 

It seems correlated quantum particles can "affect each other" instantly, even if they are in different parts of the universe. **"Spooky Action at a distance."**

The quantum state of two entangled qubits cannot be written as $ \ket{u} \otimes \ket{v} $. **" Cannot be factorized."**

# Superdense Coding (A Communication Protocol):


**Step 1: Sharing the entangled pair**\
At first, Charlie prepares the entangled pair of qubits, then:
* She sends one to Alice,
* and sends the second to Bob.

**Step 2: Message encoding by Alice**\
Alice has two bits of classical information: $ b_1,b_2 \in \{0,1\} $. 

There are four possible values for the pair $ (b_1b_2) : [ '00',\ '01',\ '10',\ '11']$. 

* If $b_1$ is 1, then Alice applies the Z-gate (Phase flip) to her qubit.

* If $b_2$ is 1, then Alice applies the X-gate (NOT operator) to her qubit.

Then, Alice sends her qubit to Bob.

<h4> After the communication </h4>

**Step 3: Message decoding by Bob**\
Bob has both qubits.

* Bob applies CX-gate (CNOT operator), where Alice's qubit is the controller.

* Then, Bob applies H-gate (Hadamard operator) to Alice's qubit.

* Bob measures both qubits. 

The measurement result will be exactly: $ b_1b_2 $.
***

### ➤ Tracing Statevector of the Quantum Circuit:

### 🧮 Step 1: Entanglement Preparation

- **Initial state**:  
  $$\ket{\psi_i} = \ket{00}$$

- **After entanglement preparation**:  
  $$\ket{\psi'} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11}) = \ket{\phi^+}$$

---

### 🧩 Step 2: Encoding by Alice

| (b_1, b_2) | Transformation     | State Sent to Bob |
|--------|---------------------|--------------------|
| 0 0    | $I \otimes I$       | $\frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$ |
| 0 1    | $X \otimes I$       | $\frac{1}{\sqrt{2}}(\ket{10} + \ket{01})$ |
| 1 0    | $Z \otimes I$       | $\frac{1}{\sqrt{2}}(\ket{00} - \ket{11})$ |
| 1 1    | $ZX \otimes I$      | $\frac{1}{\sqrt{2}}(\ket{10} - \ket{01})$ |

---

### 🔓 Step 3: Decoding and Final Measurement by Bob

| Encoded State Sent by Alice              | Apply CNOT Gate      | After CNOT | Qubit 1 | Qubit 2 | Apply $H \otimes I$ | Final State | Measurement | Classical Bits (ab) |
|----------------------------------------|-----------------------|---------------------------|---------|---------|----------------------|-------------|-------------|----------------------|
| $\frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$ | CNOT                | $$\frac{\ket{00} + \ket{10}}{\sqrt{2}}$$ | $$\frac{\ket{0} + \ket{1}}{\sqrt{2}}$$ | $$\ket{0}$$ | $H \otimes I$       | $\ket{00}$  | 00          | 00                   |
| $\frac{1}{\sqrt{2}}(\ket{10} + \ket{01})$ | CNOT                | $$\frac{\ket{11} + \ket{01}}{\sqrt{2}}$$ | $$\frac{\ket{1} + \ket{0}}{\sqrt{2}}$$ | $$\ket{1}$$ | $H \otimes I$       | $\ket{01}$  | 01          | 01                   |
| $\frac{1}{\sqrt{2}}(\ket{00} - \ket{11})$ | CNOT                | $$\frac{\ket{00} - \ket{10}}{\sqrt{2}}$$ | $$\frac{\ket{0} - \ket{1}}{\sqrt{2}}$$ | $$\ket{0}$$ | $H \otimes I$       | $\ket{10}$  | 10          | 10                   |
| $\frac{1}{\sqrt{2}}(\ket{10} - \ket{01})$ | CNOT                | $$\frac{\ket{11} - \ket{01}}{\sqrt{2}}$$ | $$\frac{\ket{1} - \ket{0}}{\sqrt{2}}$$ | $$\ket{1}$$ | $H \otimes I$       | $\ket{11}$  | 11          | 11                   |

<h3> Task 1</h3>

Verify the Superdense Coding protocol.

For each pair of $ (a,b) \in \left\{ (0,0), (0,1), (1,0),(1,1) \right\} $:
- Create a quantum circuit with two qubits: Alice's and Bob's qubits
- Both are initially set to $ \ket{0} $
- Apply h-gate (Hadamard) to Alice's qubit
- Apply cx-gate as CNOT(Alice's-qubit, Bob's-qubit)

Assume that both qubits are separated from each other.

<ul>
    <li> If $ a $ is 1, then apply z-gate to Alice's qubit. </li>
    <li> If $ b $ is 1, then apply x-gate (NOT) to Alice's qubit. </li>
</ul>

Assume that Alice sends her qubit to Bob.
- Apply cx-gate as CNOT(Alice's-qubit, Bob's-qubit)
- Apply h-gate (Hadamard) to Alice's qubit
- Measure both qubits and compare the results with pair $ (a,b) $

```python
# import all necessary objects and methods for quantum circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector

all_pairs = ['00','01','10','11']

for pair in all_pairs:

    # Create a quantum circuit with two qubits: Alice's and Bob's qubits
    # Both are initially set to |0>
    q = QuantumRegister(2, "q")    # quantum register with 2 qubits
    c = ClassicalRegister(2, "c")  # classical register with 2 bits
    qc = QuantumCircuit(q, c)      # quantum circuit with quantum and classical registers

    # Apply h-gate (Hadamard) to Alice's qubit
    qc.h(q[1])

    # Apply cx-gate as CNOT(Alice's-qubit, Bob's-qubit)
    qc.cx(q[1], q[0])

    # Display the state after entanglement
    print('Entangled State:')
    bell_state = Statevector.from_instruction(qc)
    display(bell_state.draw('latex'))

    # both qubits are separated from each other.
    qc.barrier(label=r'$|ϕ⁺⟩$')
    
    # If a is 1, then apply z-gate to Alice's qubit.
    if pair[0]=='1':
        qc.z(q[1])
        
    # If b is 1, then apply x-gate (NOT) to Alice's qubit.
    if pair[1]=='1':
        qc.x(q[1]) 

    # Alice sends her qubit to Bob.
    qc.barrier()

    # Apply cx-gate as CNOT(Alice's-qubit, Bob's-qubit)
    qc.cx(q[1], q[0])
    
    # Apply h-gate (Hadamard) to Alice's qubit
    qc.h(q[1])
    
    # Measure both qubits 
    qc.barrier()
    qc.measure(q, c)

    # draw the circuit in Qiskit's reading order
    print('\n')
    display(qc.draw(output='mpl',reverse_bits=True))

     # compare the results with pair (a,b)
    job = AerSimulator().run(qc,shots=100)
    counts = job.result().get_counts(qc)
    print(f'Input: {pair} --> Output: {counts}')
    print('-'* 150)
```

# Quantum Teleportation (A communication protocol)

Alice wants to send a qubit to Bob by using only classical communication.

Let $ \ket{\psi} = \begin{pmatrix} a \\ b \end{pmatrix} \in \mathbb{R}^2 $ be the quantum state.

_Discussion:_ If Alice has many copies of this qubit, then she can collect the statistics based on these qubits and obtain an approximation of $ a $ and $ b $, say $ \tilde{a} $ and $\tilde{b}$, respectively. After this, Alice can send $ \tilde{a} $ and $\tilde{b}$ by using many classical bits, the number of which depends on the precision of the amplitudes. 

On the other hand, if Alice and Bob share the entangled qubits in state in advance, then Bob can construct $ \ket{\psi} $ in his qubit after receiving two bits of information from Alice. 



### ➤ Tracing the State Vector Over the Circuit

### 🧮 Step 1: Entanglement Preparation

- **Initial state**:  
  $$\ket{\psi_i} = \ket{00}$$

- **After entanglement preparation**:  
  $$\ket{\psi'} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11}) = \ket{\phi^+}$$

---

### 🧩 Step 2: Encoding and Measurement by Alice

- **Alice's quantum message (key)**:
  $$
  \ket{\psi} = \begin{pmatrix} a \\ b \end{pmatrix} = a\ket{0} + b\ket{1}
  $$

Alice has two qubits, and Bob has one qubit.
- **Combined State**

The combined state after preparing the entangled pair and incorporating Alice's quantum message is:
$$
\ket{\psi_2} = (\ket{\psi} \otimes \ket{\phi^+})
$$
Expanding this:
$$
\ket{\psi_2} = (a\ket{0} + b\ket{1}) \otimes \left( \frac{1}{\sqrt{2}} (\ket{00} + \ket{11}) \right)
$$
$$
\ket{\psi_2} = \frac{1}{\sqrt{2}} \left( a\ket{000} + a\ket{011} + b\ket{100} + b\ket{111} \right)
$$

- **Applying CNOT Operator by Alice**

Applying the CNOT operator controlled on the first qubit of Alice's message and targeted by Alice's entangled qubit:
$$
\ket{\psi_3} = (\text{CNOT} \otimes I) \ket{\psi_2}
$$
$$
\ket{\psi_3} = \frac{1}{\sqrt{2}} \left( a\ket{000} + a\ket{011} + b\ket{110} + b\ket{101} \right)
$$

- **Applying Hadamard Operator by Alice**

Applying the Hadamard operator to the first qubit of Alice's message:
$$
\ket{\psi_4} = (H \otimes I \otimes I) \ket{\psi_2}
$$
$$
\ket{\psi_4} = \frac{1}{2} \left( a\ket{000} + a\ket{100} + a\ket{011} + a\ket{111} + b\ket{010} - b\ket{110} + b\ket{001} - b\ket{101} \right)
$$

Simplifying:
$$
\begin{aligned}
\ket{\psi_4} = \frac{1}{2} \space \big[ \space  &\ket{00} \space \big(a\ket{0} + b\ket{1}) \\
                                 + \space &\ket{01} \space \big(a\ket{1} + b\ket{0}\big) \\
                                 + \space &\ket{10} \space \big(a\ket{0} - b\ket{1}\big) \\
                                 + \space &\ket{11} \space \big(a\ket{1} - b\ket{0}\big) \space \big]
\end{aligned}
$$

- **Measurement by Alice**

Alice measures the first two qubits (her own qubits) and sends the result to Bob. The possible outcomes are:

| Measurement Result | Corresponding State |
|--------------------|---------------------|
| "00"               | $\ket{v_{00}} = a\ket{0} + b\ket{1}$ |
| "01"               | $\ket{v_{01}} = a\ket{1} + b\ket{0}$ |
| "10"               | $\ket{v_{10}} = a\ket{0} - b\ket{1}$ |
| "11"               | $\ket{v_{11}} = a\ket{1} - b\ket{0}$ |

Each outcome occurs with probability $\frac{1}{4}$.

---
#### 🔓 Step 3: Post-processing by Bob

Bob applies the appropriate correction based on the measurement result sent by Alice:

| Measurement Result | Correction Gate | Corrected State |
|--------------------|-----------------|------------------|
| "00"               | $I$             | $\ket{\psi} = a\ket{0} + b\ket{1}$ |
| "01"               | $X$             | $\ket{\psi} = a\ket{0} + b\ket{1}$ |
| "10"               | $Z$             | $\ket{\psi} = a\ket{0} + b\ket{1}$ |
| "11"               | $ZX$            | $\ket{\psi} = a\ket{0} + b\ket{1}$ |

Thus, regardless of the measurement outcome, Bob can reconstruct **exactly** the original state $\ket{\psi}$.

$$
\boxed{\text{Quantum teleportation successfully transfers the state } \ket{\psi} \text{ from Alice to Bob.}}
$$

<h3> Task 4: Implementing Quantum Teleportation using Qiskit </h3>
Create a quantum circuit with three qubits as described in the protocol and two classical bits.

Implement the protocol given above until Alice makes the measurements (included).
- The state of $q[2]$ can be set by the rotation with a randomly picked angle.
- Remark that Bob does not make the measurement.

At this point, read the state vector of the circuit by using "AerSimulator". 

_When a circuit having measurement is simulated by "AerSimulator", the simulator picks one of the outcomes, and so we see one of the states after the measurement._

Verify that the state of Bob' qubit is in one of these: $ \ket{v_{00}}$, $ \ket{v_{01}}$, $ \ket{v_{10}}$, and $ \ket{v_{11}}$.

Guess the measurement outcome obtained by the simulator.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.quantum_info import Statevector, random_statevector, partial_trace, DensityMatrix
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator
import numpy as np

# Step 1: Define the quantum state to teleport
state = Statevector([np.sqrt(3/4), 1/2])  # Example single-qubit state

print("Initial Quantum State:")
display(state.draw('latex', prefix="|\\psi\\rangle = "))
display(plot_bloch_multivector(state))

# Step 2: Create quantum and classical registers
qr = QuantumRegister(3, 'q')  # q2: state, q1: Alice's entangled, q0: Bob
cr = ClassicalRegister(2, 'c')  # Classical bits for Alice's measurements
qc = QuantumCircuit(qr, cr)

# Step 3: Initialize q2 with the state to be teleported
qc.initialize(state, qr[2])
# qc.x(qr[0])
# qc.x(qr[1])
qc.barrier()

# Step 4: Create Bell pair between Alice (q1) and Bob (q0)
qc.h(qr[1])
qc.cx(qr[1], qr[0])
qc.barrier()

# Step 5: Alice performs Bell-state measurement
qc.cx(qr[2], qr[1])
qc.h(qr[2])
qc.barrier(label="$|\\psi_{pre}\\rangle$")

# Save the state vector before measurement
state_pre = Statevector.from_instruction(qc)

# Step 6: Measurement
qc.measure(qr[2], cr[1])
qc.measure(qr[1], cr[0])

# Step 7: Conditional operations (classical control) on Bob's qubit (q0)
with qc.if_test((cr[0], 1)):
    qc.x(qr[0])
with qc.if_test((cr[1], 1)):
    qc.z(qr[0])

# Step 8: Save final statevector for analysis
qc.save_statevector(label="$|\\psi_{\\text{final}}\\rangle$")

# Visualize the complete circuit
display(qc.draw('mpl', reverse_bits=True, cregbundle=False))

# Step 9: Simulate the circuit
simulator = AerSimulator()

# Run the circuit
job = simulator.run(qc, shots=1)
result = job.result()
counts = result.get_counts()

print("\nPre-measurement Statevector:")
display(state_pre.draw('latex', prefix="|\\psi_{pre}\\rangle = "))
# display(plot_bloch_multivector(final_state))

# print the output measurement
print("Measurement Results:", counts)

# Step 10: Extract the final global statevector
final_state = Statevector(result.data(0)["$|\\psi_{\\text{final}}\\rangle$"])

print("\nFinal Global Statevector after measurement (Collapsed State):")
display(final_state.draw('latex', prefix="|\\psi_{\\text{final}}\\rangle = "))
# display(plot_bloch_multivector(final_state))

# Step 11: Extract Bob's qubit only (q0), trace out q1 and q2
bob_qubit = partial_trace(final_state, [1, 2])

print("Bob's Qubit Final State:")
display(bob_qubit.to_statevector().draw('latex', prefix="|\\psi_{Bob}\\rangle = "))
display(plot_bloch_multivector(bob_qubit))

# Step 12: Verify fidelity (optional)
from qiskit.quantum_info import state_fidelity

# Convert the original state to DensityMatrix to match types
original_dm = DensityMatrix(state)

# Compute fidelity between Bob's qubit and the original state
fidelity_value = state_fidelity(bob_qubit, original_dm)

print(f"Fidelity between original and teleported state: {fidelity_value:.6f}")
```
